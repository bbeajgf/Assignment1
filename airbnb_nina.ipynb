GET RID MISSING VALUES!

```python
airbnb.drop(['name', 'host_id', 'host_name', 'neighbourhood_group', 'neighbourhood'], axis=1, inplace=True)
airbnb
```

```python
# fill the last_review with the most common date
airbnb["last_review"] = airbnb["last_review"].fillna(airbnb["last_review"].mode().iloc[0])
airbnb
```

```python
# fill missing value in reviews_per_month with the mean
airbnb["reviews_per_month"] = airbnb["reviews_per_month"].fillna(airbnb["reviews_per_month"].mean())
airbnb
```

```python
airbnb.isnull().sum()
```

New Category based on price

```python
airbnb["price_cat"] = np.where(airbnb["price"] > 500, 'Expensive', 
                            (np.where(airbnb["price"] > 250, 'Medium', 'Cheap')))
airbnb
```

Filter last_review upon 2019

```python
# filter last_review > 2019
airbnb.loc[airbnb["last_review"] > '2019']
```

Price Visualization

```python
room_type = airbnb["room_type"]
price = airbnb["price"]

plt.bar(room_type, price)
plt.title('Price of Available Room', loc='center', fontsize=30, color='blue')
plt.xlabel('Room Type', fontsize=15)
plt.ylabel('Price', fontsize=15)

plt.tight_layout()
plt.savefig('graph.png')
plt.show()
```
