### Exercise 4

Find all trips that is greater than 10 miles and originated from New York and Morrisville

```python
trips = uber.loc[uber["MILES*"] > 10]
trips
```

```python
trips1 = uber.loc[uber["START*"].isin(["New York","Morrisville"])]
trips1
```

```python
trip = uber.loc[(uber["MILES*"] > 10) & (uber["START*"].isin(["New York","Morrisville"]))]
trip
```

### Exercise 5

Create a new column with the following condition:
a) >10 : Long Trip
b) 5-10 : Medium Trip
c) <5 : Short Trip

```python
import numpy as np
```

```python
uber["DIST_CAT"] = np.where(uber["MILES*"] > 10, "Long Trip", 
                            (np.where(uber["MILES*"] > 5, "Medium Trip", "Short Trip")))
uber
```
